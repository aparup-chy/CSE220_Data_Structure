```python

class Patient:
    def __init__(self,id,name,age,blgroup):
        self.id=id
        self.name=name
        self.age=age
        self.blgroup=blgroup
        self.next=None
        self.prev=None
        
class WRM:
    def __init__(self):
        self.head=Patient(None,None,None,None)
        self.head.next=self.head
        self.head.prev=self.head

    def registerPatient(self,id,name,age,blgroup):
        nwpatient=Patient(id,name,age,blgroup)
        if self.head.next==self.head:
            nwpatient.next=self.head
            nwpatient.prev=self.head
            self.head.next=nwpatient
            self.head.prev=nwpatient
        else:
            lastpatient=self.head.prev
            nwpatient.next=self.head
            nwpatient.prev=lastpatient
            self.head.prev=nwpatient
            lastpatient.next=nwpatient

    def patentservice(self):
        if self.head.next==self.head:
            print("No patients in the waiting room.")
            return
        else:
            servedpatient=self.head.next
            self.head.next=servedpatient.next
            servedpatient.next.prev=self.head
            print("Serving patient:",servedpatient.name)

    def cancel(self):
        self.head.next=self.head
        self.head.prev=self.head
        print("All appointments canceled. Doctor can go to lunch.")

    def doctorgoinghome(self):
        if self.head.next==self.head:
            return True
        else:
            return False

    def patientlist(self):
        if self.head.next==self.head:
            print("No patients in the waiting room.")
            return
        else:
            ongoingpatients=self.head.next
            while ongoingpatients!=self.head:
                print("Patient ID:",ongoingpatients.id)
                ongoingpatients=ongoingpatients.next

#Tester
wrm=WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")

    choice=int(input("Enter your choice: "))
    if choice==1:
        id=int(input("Enter patient ID: "))
        name=input("Enter patient name: ")
        age=int(input("Enter patient age: "))
        blgroup=input("Enter patient blood group: ")
        wrm.registerPatient(id,name,age,blgroup)
        print("Patient added successfully.")
    elif choice==2:
        wrm.patentservice()
    elif choice==3:
        wrm.patientlist()
    elif choice==4:
        if wrm.doctorgoinghome():
            print("Yes, doctor can go home.")
        else:
            print("No, there are patients waiting.")
    elif choice==5:
        wrm.cancel()
    elif choice==6:
        break
    else:
        print("Invalid choice. Please try again.")
        
```
